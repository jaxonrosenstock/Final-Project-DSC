## **Types of Machine Learning**


### 1. **Supervised Learning**

Supervised learning is a machine learning process in which both input and desired output data are provided. The purpose of supervised learning is to find a function that can be used to assign a class or value to an input value. For this purpose, the data is provided with a so-called label.

**Types of Supervised Learning**

- **Regressions**: If the desired output consists of one or more continuous variables, then the task is called regression. An example of a regression problem would be the predicting housing price based on the input features of previously sold houses.

- **Classifications**: In this process samples belong to two or more classes and we want to learn from already labeled data how to predict the class of unlabeled data. An example of a classification problem would be handwritten digit recognition, in which the aim is to assign each input vector to one of a finite number of discrete categories.

**Reference Video**: [Supervised Learning](https://www.youtube.com/watch?v=bQI5uDxrFfA)

### 2. **Unsupervised Learning**

The aim of the unsupervised learning approach is to recognize unknown patterns from the data and derive rules from them. In this process, the training data consists of a set of input vectors x without any corresponding target values. The goal in such problems may be to discover groups of similar examples within the data.
- **Clustering**:  Clustering is the task of grouping a set of objects in such a way that objects in the same group (called a cluster) are more similar (in some sense) to each other than to those in other groups (clusters)(Wiki).
- **Artificial Neural Networks**: Artificial neural networks (ANN) are computing systems vaguely inspired by the biological neural networks that constitute animal brains. The neural network itself isn't an algorithm, but rather a framework for many different machine learning algorithms (both supvervised and unsupervised) to work together and process complex data inputs (Wiki).

**Reference Video**: [Unsupervised Learning](https://www.youtube.com/watch?v=jAA2g9ItoAc)


### 3. **Reinforcement Learning**

Reinforcement learning is a type of machine learning where an agent learns to make decisions in an environment by interacting with it and receiving feedback in the form of rewards or punishments. The goal of reinforcement learning is to learn an optimal policy, which is a set of actions that maximizes the expected cumulative reward over time.

In reinforcement learning, the agent takes an action in the environment, which transitions it to a new state and provides a reward signal. The agent then uses this information to update its policy to take better actions in the future. This process is repeated over many episodes, and the agent learns from experience to take actions that lead to the highest cumulative reward.

Reinforcement learning has several applications, including:

1. Game playing: Reinforcement learning has been used to train agents to play games such as chess, Go, and poker.

2. Robotics: Reinforcement learning is used in robotics to train agents to perform tasks such as object manipulation and navigation.

3. Control systems: Reinforcement learning can be used to optimize control systems in manufacturing and other industrial settings.

4. Recommendation systems: Reinforcement learning can be used to personalize recommendations for users based on their interactions with a system.

5. Finance: Reinforcement learning can be used to optimize trading strategies in financial markets.

## **Components of Machine Learning Models**

### **1. Input data preprocessing**

**Data Cleaning:**

- Data cleaning involves identifying and correcting errors or inconsistencies in the dataset. This may include:

    - **Handling missing values**: Imputation techniques such as mean, median, or mode imputation.

    - **Removing duplicate records**: Identifying and removing duplicate entries to avoid bias in the analysis.
    
    - **Dealing with outliers**: Detecting and treating outliers using statistical methods or domain knowledge.

**Data Transformation:**

Data transformation involves converting the dataset into a suitable format for analysis. This may include:

- **Encoding categorical variables**: Converting categorical variables into numerical format that can be used by machine learning algorithms. This may involve one-hot encoding, label encoding, or ordinal encoding depending on the nature of the categorical variables.

- **Feature engineering**: Creating new features from existing ones or transforming features to capture relevant information for the model.


**Data Integration:**

Data integration involves combining multiple datasets or data sources into a unified dataset for analysis. This may include merging datasets based on common identifiers or joining tables based on specific relationships.

### **2. Train/Test Split**

The train-test split is a fundamental technique used in machine learning to train and then evaluate the performance of a model on unseen data. The basic rule for the train-test split is to partition the available dataset into two subsets: one for training the model and the other for testing the model's performance.


**a. Training Set:**

The training set typically comprises a larger portion of the dataset, often around 70-85% of the total data.
The model learns patterns and relationships in the training data to make predictions.


**b. Testing Set:**

The testing set is kept separate from the training set and is not used during the model training phase.
It serves as an independent dataset to assess the model's performance and generalization ability on unseen data.
The testing set usually contains the remaining 15-30% of the data.

### **3. Data normalization/standardization**


Data normalization is an essential preprocessing step in machine learning to ensure that all features have a similar scale and distribution. This step helps improve the convergence of optimization algorithms, prevents certain features from dominating others, and generally leads to better performance of machine learning models.


  - **Min-Max Scaling:** Scales the features to a fixed range, typically between 0 and 1.
  - **Standardization (Z-score Normalization)**: Transforms the features to have a mean of 0 and a standard deviation of 1.
  - **Log Transformation**: Applies a logarithmic transformation to skewed features to make their distribution more symmetrical.


**Handling Categorical Features:**

- **Categorical features** may require special treatment during normalization, depending on the encoding method used.
One-hot encoding often results in binary features, which do not require normalization.
- **For ordinal encoding**, normalization can be applied similarly to numerical features.

### **4. Model building/selection**

Choose an appropriate machine learning algorithm based on the problem type (classification, regression, clustering, etc.), the nature of the data, and the desired outcome.

**Example: Parameter Optimization in Linear Regression**

 Model Equation: $\hat{y} = wx + b$

 - Goals: estimate $w$ and $b$ (model paramaters)

- $\hat{y}$: the predicted value
- $x$: the input feature
- $w$: the weight (slope)
- $b$: the bias (intercept)

### **5. Model training**

Fit the selected model to the training data by feeding it the input features and corresponding target labels.
During training, the model learns the underlying patterns and relationships in the data, adjusting its parameters to minimize the prediction error.

#### **How does a machine learn the given task?**

**a. Parameter Initialization:**

Parameters (coefficients) of the regression model are initialized with random values or zeros.

**b. Forward Pass:**

The model takes input features and makes predictions based on the current parameter values.
For linear regression, the predicted values are calculated as the dot product of input features and model parameters, possibly with an additional intercept term.

**c. Loss Calculation:**

The loss function is calculated to quantify the discrepancy between the predicted values and the actual target values in the training data.
In regression problems, common loss functions include Mean Squared Error (MSE).

We aim to minimize the Mean Squared Error (MSE) between the predicted and actual values:


$\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 = \frac{1}{n} \sum_{i=1}^{n} (y_i - (w x_i + b))^2$


**d.Gradient Calculation:**

The gradient of the loss function with respect to each parameter is calculated using techniques like backpropagation or analytic differentiation.
The gradient represents the direction and magnitude of the steepest ascent or descent of the loss function.

- Initialize $w$ and $b$ randomly
- Calculate the MSE loss
- Compute gradients:
    
    - $$\frac{\partial \text{MSE}}{\partial w} = -\frac{2}{n} \sum_{i=1}^{n} x_i (y_i - \hat{y}_i)$$

    - $$\frac{\partial \text{MSE}}{\partial b} = -\frac{2}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)$$

**e. Parameter Update:**

The parameters are updated in the opposite direction of the gradient to minimize the loss function.
Gradient descent algorithms adjust the parameters iteratively by taking steps proportional to the negative gradient.
The learning rate (step size) determines the size of each parameter update and is a hyperparameter that needs to be tuned.

- Update parameters (Gradient Descent):

    $$w := w - \alpha \cdot \frac{\partial \text{MSE}}{\partial w}$$

    $$b := b - \alpha \cdot \frac{\partial \text{MSE}}{\partial b}$$

    where $\alpha$ is the learning rate.

**f. Iteration:**

Steps 2-5 are repeated iteratively for a predefined number of epochs or until convergence criteria are met.
During each iteration, the model gradually learns from the training data and updates its parameters to minimize the loss function.

**g. Convergence:**

Convergence occurs when the optimization algorithm reaches a point where further parameter updates do not significantly reduce the loss function.
At convergence, the model has found parameter values that approximate the optimal solution for the regression problem.

### **6. Model prediction**

Once a machine learning model has been trained on a dataset (i.e., it has learned patterns and relationships within the data), it can be deployed to make predictions on new data instances (test data).

**Final model with updated $w$ and $b$**

$$ y = wx+b$$

### **7. Model's perforamnce evaluation**

The predictions made by the model need to be evaluated to assess their accuracy and reliability. This involves comparing the predicted values with ground truth labels or actual outcomes to measure the model's performance.

Performance evaluation metrics for regression.

- **Mean Squared Error (MSE)**

    $$\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$$


- **Root Mean Squared Error (RMSE)**
   
    $$\text{RMSE} = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2}$$


- **Mean Absolute Error (MAE)**
    
    $$\text{MAE} = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|$$


- **Mean Absolute Percentage Error (MAPE)**

    $$\text{MAPE} = \frac{1}{n} \sum_{i=1}^{n} \left| \frac{y_i - \hat{y}_i}{y_i} \right|\times 100\% \quad \text{(requires } y_i \neq 0\text{)}$$


- **R-squared ($R^2\) – Coefficient of Determination**

    $$R^2 = 1 - \frac{\sum_{i=1}^{n} (y_i - \hat{y}_i)^2}{\sum_{i=1}^{n} (y_i - \bar{y})^2}$$



- **Adjusted R-squared**

    $$R_{\text{adj}}^2 = 1 - \frac{n - 1}{n - p - 1}\times \left(1 - R^2\right)$$
    where $p$ is the number of predictors in the model.
